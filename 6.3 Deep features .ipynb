```python
import graphlab
```

```python
#load the CIFR-10 Data set
image_train = graphlab.SFrame('image_train_data1/')
```

```python
image_train.head()
```

```python
#Train nearest model using deep feathures
knn_model = graphlab.nearest_neighbors.create(image_train,features=['deep_features'],label='id')
```

```python
graphlab.canvas.set_target('ipynb')
cat = image_train[18:19]
cat.show()
```

```python
knn_model.query(cat)
```

```python
def get_images(query_res):
    return image_train.filter_by(query_res['reference_label'],'id')
```

```python
cat_nea = get_images(knn_model.query(cat))
```

```python
cat_nea['image'].show()
```

```python
car = image_train[8:9]
car.show()
```

```python
get_images(knn_model.query(car))['image'].show()
```
